```python
from dataclasses import dataclass

@dataclass
class User:
    name: str
    age: int

u = User(name="Rahul", age="25")  # age stays string ‚ùå
```

```python

from pydantic import BaseModel
```

```python
class Person(BaseModel):
    name:str
    age:int
    city:str

person=Person(name="Rahul",age=20,city="sonipat")
```
